```python
import numpy as np
import pandas as pd
```

```python
moviesData = pd.read_csv('tmdb_5000_movies.csv')
creditsData = pd.read_csv('tmdb_5000_credits.csv')
```

```python
moviesData.head(1)
```

```python
creditsData.head(1)
```

```python
# As we can see there are 2 DF we can also merge them to have one single database df

movies = moviesData.merge(creditsData,on= 'title')
```

```python
movies.head()
```

```python
# As there are multiple columns so now lets filter out the columns that will be useful to create the content recommendation system (using tags)
# Columns we can keep for content based recommendation system
'''
Genres 
id
Keywords
Title
Overview
Cast
Crew
'''
movies = movies[['genres','id','keywords','title','overview','cast','crew']]
```

```python
# So now our dataset df is lookiong like this
movies.head()
```

```python
# Now as we have to do content based recomm we will merge few columns
'''
id
title
tags --> merge keywords,overview,top 3 cast & director from crew.    (USED FOR CONTENT BASED RECOM SYSTEM)
'''

# DATA PREPROCESSING

# 1- Missing data ko fix karna

movies.isna().sum()  #we can see we have 3 nan values in overview so drop them
```

```python
movies.dropna(inplace=True)  #Sari nan values ko nikal dia
```

```python
movies.isna().sum()
```

```python
# 2 - Check duplicate data

movies.duplicated().sum()   #we have 0 duplicated data
```

```python
movies.loc[0].genres
```

```python
# As we get dict when we see genres we just need name eg - action , adventure , fantasy etc 
import ast
def gen(obj):
    l1 = []
    for i in ast.literal_eval(obj):
        l1.append(i['name'])
    return l1


```

```python
movies['genres'] = movies['genres'].apply(gen)
```

```python
# Do the same thing for keywords as it also have same structure dict and we need to extract names only

movies['keywords'] = movies['keywords'].apply(gen)
```

```python
# Lets do for cast abhi as we only neew top 3 cast mem so lets create a fn to extract it
def cas(obj):
    l1 = []
    count = 0
    for i in ast.literal_eval(obj):
        if count != 3:
            l1.append(i['name'])
            count += 1
        else:
            break
    return l1


movies['cast'] = movies['cast'].apply(cas)
```

```python
movies.head()
```

```python
# Abhi yeh same cheez crew ke lia karna hai
movies.loc[1].crew
```

```python
# as we only need directiors

def getDirector(obj):
    director = []

    for i in ast.literal_eval(obj):
        if i['job'] == 'Director':
            director.append(i['name'])
            return director
    return ''
```

```python
movies['crew'] = movies['crew'].apply(getDirector)
```

```python
movies.head()
```

```python
movies.tail()
```

```python
# As overview is in string format so we will convert it into list of words
movies['overview'] = movies['overview'].apply(lambda x: x.split())
```

```python
movies.head()
```

```python
# so now as we have converted everything into a list then abhi fir sab names ke beech mai se space remove krke single word banana hai
# eg - Sam Worthington  --> SamWorthington.  to not to confuse the model with just first name 

movies['genres'] = movies['genres'].apply(lambda x: [i.replace(" ",'') for i in x])
```

```python
movies['keywords'] = movies['keywords'].apply(lambda x: [i.replace(" ",'') for i in x])
movies['cast'] = movies['cast'].apply(lambda x: [i.replace(" ",'') for i in x])
movies['crew'] = movies['crew'].apply(lambda x: [i.replace(" ",'') for i in x])
```

```python
movies.head()
```

```python
movies['Tags'] = movies['overview'] + movies['genres'] + movies['cast'] + movies['crew']
```

```python
movies.head()
```

```python
dataset = movies[['id','title','Tags']]
```

```python
dataset
```

```python
# Now as tags is a list we have to convery it into a string
dataset['Tags'] = dataset['Tags'].apply(lambda x: " ".join(x))
```

```python
dataset['Tags'][0]
```

```python
# Now lets convert it to lower case
dataset['Tags'] = dataset['Tags'].apply(lambda x: x.lower())
```
